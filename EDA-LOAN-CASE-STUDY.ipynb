```python
# Import all relevant Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

**Load CSV data on Panda Dataframe**

```python
# Loading lean.csv in panda dataframe
df_orig_loan = pd.read_csv('/content/drive/MyDrive/Data-For-Analysis/loan.csv')
df_orig_loan.head()
```

```python
# Analysis of Dataframe
print("Dimension of df",df_orig_loan.shape)
print("\n")
print("**Column Info********************************************************")
info_loan=df_orig_loan.info(verbose=True)
print("Dataframe column details", "\n",info_loan)
```

Get Details of Dataframe

---



```python
# Increasing the row count of display
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 50)

# Analysis of Data frameColumn with null count
null_col_count=df_orig_loan.isnull().sum()
print(null_col_count.shape)
print("Dataframe column having null values", "\n",null_col_count[null_col_count>0])
print("********************************************************************")
print("Count of column having all Null rows ", null_col_count[null_col_count == df_orig_loan.shape[0]].count())
```

```python
# Cleaning the dataframe Dropping the columns having all null values
df_no_col_NA_loan=df_orig_loan.drop(null_col_count[null_col_count == df_orig_loan.shape[0]].index, axis=1)
null_col_count_after=df_no_col_NA_loan.isnull().sum()

print("Dataframe column having null values", "\n",null_col_count_after[null_col_count_after>0])
```

```python
# Dropping the columns having majority of null values
#mths_since_last_delinq [25682],mths_since_last_record[36931] and next_pymnt_d[38577]
df_no_col_NA_loan=df_no_col_NA_loan.drop(['mths_since_last_delinq','mths_since_last_record','next_pymnt_d'], axis=1)

df_no_col_NA_loan.shape

null_col_count_after=df_no_col_NA_loan.isnull().sum()

print("Dataframe column having null values", "\n",null_col_count_after[null_col_count_after>0])
```

```python
# Dimension of dataframe
print("Dimension of dataframe",df_no_col_NA_loan.shape)
# Columns not needes for analysis are emp_title, desc
df_no_col_NA_loan=df_no_col_NA_loan.drop(['emp_title','desc'], axis=1)

null_col_count_after=df_no_col_NA_loan.isnull().sum()

print("Dataframe column having null values", "\n",null_col_count_after[null_col_count_after>0])
print("Dimension of dataframe",df_no_col_NA_loan.shape)
```

```python
# Cleaning of dataframe

# pub_rec_bankruptcies [Number of public record bankruptcies], changing NA to 0
df_no_col_NA_loan['pub_rec_bankruptcies'].fillna(0, inplace=True)

#Drop pymnt_plan as it has only one value
df_no_col_NA_loan=df_no_col_NA_loan.drop(['pymnt_plan'], axis=1)

# Drop url column as not needed for analysis
df_no_col_NA_loan=df_no_col_NA_loan.drop(['url'], axis=1)

# Drop desc col as not needed for analysis
#df_no_col_NA_loan=df_no_col_NA_loan.drop(['desc'], axis=1)

# Drop title col as not needed for analysis
df_no_col_NA_loan=df_no_col_NA_loan.drop(['title'], axis=1)

#Drop initial_list_status as it has only one value
df_no_col_NA_loan=df_no_col_NA_loan.drop(['initial_list_status'], axis=1)

# Drop aaplication_type as it has only one value individual
df_no_col_NA_loan=df_no_col_NA_loan.drop(['application_type'], axis=1)

# Dropping policy_code,collections_12_mths_ex_med,acc_now_delinq,chargeoff_within_12_mths,delinq_amnt and tax_liens as has only one value
df_no_col_NA_loan=df_no_col_NA_loan.drop(['policy_code','collections_12_mths_ex_med','acc_now_delinq','chargeoff_within_12_mths','delinq_amnt','tax_liens'], axis=1)

null_col_count_after=df_no_col_NA_loan.isnull().sum()

print("Dataframe column having null values", "\n",null_col_count_after[null_col_count_after>0])
```

```python
# Dimension of dataframe
print("Dimension of dataframe",df_no_col_NA_loan.shape)

print("**Column Info********************************************************")
info_loan=df_no_col_NA_loan.info(verbose=True)
print("Dataframe column details", "\n",info_loan)

object_columns = df_no_col_NA_loan.select_dtypes(include='object').columns
print("*********************************************************************","\n")
print("Object data tupe columns ",object_columns)
print("Object data tupe columns count",len(object_columns))
```

```python
# Cleaning dataframe Correcting column variable type
# term converting to int after removing months
df_no_col_NA_loan['term_in_month']= df_no_col_NA_loan['term'].str.replace(' months','')
df_no_col_NA_loan['term_in_month']= df_no_col_NA_loan['term_in_month'].astype(int)
df_no_col_NA_loan.drop(['term'], axis=1, inplace=True)

# int_rate column converting to percentage
df_no_col_NA_loan['int_rate_percent']= df_no_col_NA_loan['int_rate'].str.replace('%','')
df_no_col_NA_loan['int_rate_percent']= df_no_col_NA_loan['int_rate_percent'].astype(float)
df_no_col_NA_loan.drop(['int_rate'], axis=1, inplace=True)


# Converting issue_d as date in mmm-yy format
df_no_col_NA_loan['issue_d']= pd.to_datetime(df_no_col_NA_loan['issue_d'], format='%b-%y')

# Converting earliest_cr_line as date in mmm-yy format
df_no_col_NA_loan['earliest_cr_line']= pd.to_datetime(df_no_col_NA_loan['earliest_cr_line'], format='%b-%y')

# revol_util column converting to percentage
df_no_col_NA_loan['revol_util_percent']= df_no_col_NA_loan['revol_util'].str.replace('%','')
df_no_col_NA_loan['revol_util_percent']= df_no_col_NA_loan['revol_util_percent'].astype(float)
df_no_col_NA_loan.drop(['revol_util'], axis=1, inplace=True)

# Converting last_pymnt_d as date in mmm-yy format
df_no_col_NA_loan['last_pymnt_d']= pd.to_datetime(df_no_col_NA_loan['last_pymnt_d'], format='%b-%y')

# Converting last_credit_pull_d as date in mmm-yy format
df_no_col_NA_loan['last_credit_pull_d']= pd.to_datetime(df_no_col_NA_loan['last_credit_pull_d'], format='%b-%y')




```

```python
# Dimension of dataframe
print("Dimension of dataframe",df_no_col_NA_loan.shape)

print("Dataframe column having null values", "\n",null_col_count_after[null_col_count_after>0])

print("**Column Info********************************************************")
info_loan=df_no_col_NA_loan.info(verbose=True)
print("Dataframe column details", "\n",info_loan)

object_columns = df_no_col_NA_loan.select_dtypes(include='object').columns
print("*********************************************************************","\n")
print("Object data tupe columns ",object_columns)
print("Object data tupe columns count",len(object_columns))

#Creating Final DataFrame
df_final_loan=df_no_col_NA_loan.copy()
```

```python
# Remove ID and Member ID
df_final_loan_1 =df_final_loan.drop(['id','member_id'], axis=1)
# Split Data frame based on Loan Status
df_final_loan_1_Charged_off =df_final_loan_1[df_final_loan_1['loan_status']=='Charged Off']
df_final_loan_1_Fully_Paid =df_final_loan_1[df_final_loan_1['loan_status']=='Fully Paid']
df_final_loan_1_Current =df_final_loan_1[df_final_loan_1['loan_status']=='Current']

```
